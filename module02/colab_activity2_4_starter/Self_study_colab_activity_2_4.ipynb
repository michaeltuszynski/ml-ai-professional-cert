### Problem 1



Given the scatterplot below, determine whether the variables `x1` and `y` are positively, negatively, or not correlated.

Assign your solution as a string -- either `positive`, `negative`, or `not` to `ans_1` below.


```python
import matplotlib.pyplot as plt
import cv2

def show_image_matplotlib(image_path):
    img = cv2.imread(image_path)
    if img is None:
        print(f"Error: Could not read image at {image_path}")
        return

    # Convert BGR to RGB (OpenCV uses BGR by default)
    img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

    plt.figure(figsize=(10, 8))
    plt.imshow(img_rgb)
    plt.axis('off')
    plt.show()
```

```python
show_image_matplotlib(os.getcwd() + "/images/guess_1.png")
```

```python
ans1 = "positive"
```
