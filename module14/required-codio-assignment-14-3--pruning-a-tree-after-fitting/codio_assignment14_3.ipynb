```python
### GRADED

train_accs = []
test_accs = []
for i in ccp_alphas[:-1]:
    pass


# YOUR CODE HERE
train_accs = []
test_accs = []
for i in ccp_alphas[:-1]:  # Using [:-1] as shown in visualization code
    # Create and fit tree with current alpha
    dtree = DecisionTreeClassifier(random_state=42, ccp_alpha=i)
    dtree.fit(X_train, y_train)

    # Calculate and store accuracies
    train_acc = dtree.score(X_train, y_train)
    test_acc = dtree.score(X_test, y_test)

    train_accs.append(train_acc)
    test_accs.append(test_acc)

# # Answer check
plt.step(ccp_alphas[:-2], train_accs[:-1], '--o', label = 'Train')
plt.step(ccp_alphas[:-2], test_accs[:-1], '--o', label = 'Test')
plt.plot(ccp_alphas[np.argmax(test_accs)], max(test_accs), 'ro', markersize = 12, alpha = 0.4, label = 'Best Accuracy')
plt.legend()
plt.grid()
plt.xlabel('alpha')
plt.ylabel('Accuracy');
```

For more examples of using Cost Complexity Pruning in scikitlearn see the example documentation [here](https://scikit-learn.org/stable/auto_examples/tree/plot_cost_complexity_pruning.html), and for more details on how the algorithm determines where to prune and how to compute alpha see [here](http://mlwiki.org/index.php/Cost-Complexity_Pruning).  

```python
fig, ax = plt.subplots(1,2, figsize = (30, 10))
plot_tree(DecisionTreeClassifier(random_state=42).fit(X_train, y_train), ax = ax[0], filled = True, feature_names=X.columns);
ax[0].set_title('Unpruned Tree', fontsize = 20)
plot_tree(DecisionTreeClassifier(random_state = 42, ccp_alpha=ccp_alphas[np.argmax(test_accs)]).fit(X_train, y_train), filled = True, feature_names = X.columns);
ax[1].set_title('Pruned Tree', fontsize = 20);
```

### Key Takeaways

1. **Pruning Mechanics**: Cost complexity pruning provides a systematic way to reduce tree complexity by adjusting the `ccp_alpha` parameter.

2. **Complexity Tradeoffs**:
   - Higher alpha values lead to simpler trees (fewer nodes and lower depth)
   - Zero alpha results in fully grown trees with potential overfitting
   - There exists an optimal alpha that balances model complexity and performance

3. **Performance Impact**:
   - Training accuracy generally decreases as alpha increases (tree simplifies)
   - Test accuracy often peaks at an intermediate alpha value, showing pruning's effectiveness in reducing overfitting

4. **Visualization**: The notebook demonstrates how to visualize:
   - Impurity vs. alpha relationship
   - Node count vs. alpha relationship
   - Tree depth vs. alpha relationship
   - Training/test accuracy vs. alpha relationship
