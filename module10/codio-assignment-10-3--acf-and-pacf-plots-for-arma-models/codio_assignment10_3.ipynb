### Problem 1

#### Is it Stationary? 

**10 Points**

As discussed, our ARMA models are only applicable to stationary data.  Use the `adfuller` function to determine if the recruits data is stationary at the 0.05 level.  Assign your answer as a string to `ans1` below.  

```python
### GRADED

ans1 = ''

# YOUR CODE HERE
# Perform Augmented Dickey-Fuller test
result = adfuller(recruits['value'])

# Check if p-value < 0.05 to determine stationarity
ans1 = 'yes' if result[1] < 0.05 else 'no'

# Answer check
print(ans1)
print(type(ans1))
```

### Problem 3

#### Examining acf and pacf

**10 Points**

Below, the ACF and PACF plots are shown.  While the ACF plot isn't incredibly helpful, the PACF may suggest using a value of `p = 1` in an ARMA model.  As such, create and fit an ARIMA model with `p = 1` and `q = 1`.  Assign your fit model as `arma` below.

```python
fig, ax = plt.subplots(1, 2, figsize = (16, 5))
plot_acf(y_hist, ax = ax[0]);
ax[0].grid()
plot_pacf(y_hist, ax = ax[1], method = 'ywm');
ax[1].grid()
```

```python
y_hist.index = pd.to_datetime(y_hist.index)
```

```python
y_hist.info()
```

```python
### GRADED

arma = ''

# YOUR CODE HERE
# Create and fit ARIMA model with p=1, q=1
arma = ARIMA(y_hist, order=(1, 0, 1)).fit()

# Answer check
print(arma)
```

### Problem 6

#### The GNP series

**10 Points**

Below, the `gnp` data is loaded and displayed.  This data is transformed according to the first difference of the logarithm so as to form a stationary series.  Then, the ACF and PACF plots are shown on the stationary series.  These suggest that an AR(2) and MA(2) model might be appropriate.  Build an `ARIMA` model on `y` and predict as `preds`.  Uncomment the code to visualize the predictions.

```python
gnp = pd.read_csv('data/gnp.csv', index_col=0)
gnp.index = pd.Index(pd.date_range("1947-Q1", "2002-Q4", freq = "Q"))
gnp.head()
```

```python
gnp.plot()
```

```python
y = np.log(gnp).diff().dropna()
```

```python
#note the stationarity
adfuller(y)
```

```python
fig, ax = plt.subplots(1, 2, figsize = (14, 4))
plot_acf(y, ax = ax[0]);
ax[0].grid()
plot_pacf(y, ax = ax[1])
ax[1].grid();
```

```python
### GRADED

arma2 = ''
preds = ''

# YOUR CODE HERE
arma2 = ARIMA(y, order=(2, 0, 2)).fit()
preds = arma2.predict()

# # Answer check
plt.figure(figsize = (12, 4))
plt.plot(y, label = 'data')
plt.plot(preds, label = 'predictions')
plt.legend()
plt.grid()
plt.title('Result of ARMA Model');
```

### Key Takeaways

1. **Stationarity is Essential**: Before applying ARMA models, data must be tested for stationarity (using ADF test)
2. **Model Selection**: ACF and PACF plots guide parameter selection:
   - ACF shows MA(q) components
   - PACF shows AR(p) components
3. **Data Transformation**: Non-stationary data (like GNP) can be transformed using techniques like log differencing
4. **Model Validation**: Residual analysis helps verify if the model has captured all temporal dependencies
5. **Time Series Splitting**: Unlike regular data, time series requires temporal ordering preservation (no shuffling)
6. **Forecasting Capabilities**: ARMA models can both predict within training data and forecast future values
