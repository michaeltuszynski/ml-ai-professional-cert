```python
import pandas as pd
import plotly.express as px
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
gapminder = px.data.gapminder()
```

### Problem 1

#### Histogram of Life Expectancy with `plotly`



Create a histogram of the Life Expectancy data using `plotly`.

```python

#make a histogram of life expectancy in plotly
fig = px.histogram(gapminder, x="lifeExp", title="Life Expectancy Distribution")
fig.show()
```

### Problem 2

#### Histogram of Life Expectancy with `seaborn`



Create a histogram of the Life Expectancy data using `seaborn`. 

```python

#make a histogram in seaborn
plt.figure(figsize=(10, 6))
sns.histplot(data=gapminder, x="lifeExp")
plt.title("Life Expectancy Distribution")
plt.show()
```

### Problem 3

#### Histogram and KDE of Life Expectancy with `seaborn` 



Create a histogram with the KDE of the Life Expectancy data using `seaborn`.

```python

#make a kde plot in seaborn
plt.figure(figsize=(10, 6))
sns.histplot(data=gapminder, x="lifeExp", kde=True)
plt.title("Life Expectancy Distribution with KDE")
plt.show()
```

### Problem 4

#### Histogram, KDE, and rugplot of Life Expectancy with `seaborn`



Create a histogram and KDE of the Life Expectancy data along with the rugplot using `seaborn`.

```python

#make a rug plot with seaborn
plt.figure(figsize=(10, 6))
sns.histplot(data=gapminder, x="lifeExp", kde=True)
sns.rugplot(data=gapminder, x="lifeExp")
plt.title("Life Expectancy Distribution with KDE and Rug Plot")
plt.show()
```

### Problem 11

#### Scatterplot of Life Expectancy vs. GDP including marginal histograms with `plotly`



Create a Scatterplot Life Expectancy vs. GDP including marginals as histograms using `plotly`.

```python

#plotly gdp v life exp marginal_x = "histogram" and y
fig = px.scatter(gapminder, x="gdpPercap", y="lifeExp",
                 marginal_x="histogram", marginal_y="histogram",
                 title="Life Expectancy vs GDP")
fig.show()
```

[Back to top](#Index:) 

### Problem 12

#### Scatterplot of Life Expectancy vs. GDP including marginal box plots with `plotly`



Create a Scatterplot Life Expectancy vs. GDP including marginals as box plots using `plotly`.

```python

#marginals with boxplot
fig = px.scatter(gapminder, x="gdpPercap", y="lifeExp",
                 marginal_x="box", marginal_y="box",
                 title="Life Expectancy vs GDP")
fig.show()
```

[Back to top](#Index:) 

### Problem 13

#### Density Heatmap of Life Expectancy vs. GDP including marginal histograms with `plotly`



Create a Density Heatmap of Life Expectancy vs. GDP including marginals as histograms using `plotly`. 

```python

#density heatmap with plotly
fig = px.density_heatmap(gapminder, x="gdpPercap", y="lifeExp",
                        marginal_x="histogram", marginal_y="histogram",
                        title="Life Expectancy vs GDP Density")
fig.show()
```

[Back to top](#Index:) 

### Problem 14

#### `jointplot` of Life Expectancy vs. GDP with `kind = hex` using `seaborn`



Create a `jointplot` of Life Expectancy vs. GDP and set the `kind` argument to `hex`. 

```python

#seaborn jointplot with kind = "hex"
sns.jointplot(data=gapminder, x="gdpPercap", y="lifeExp", kind="hex")
plt.show()
```

### Key Takeaways

1. **Multiple Perspectives**: Different visualization types reveal different aspects of the same data. For example:
   - Histograms show overall distribution
   - Violin plots reveal distribution shape and density
   - Box plots highlight statistical summaries and outliers

2. **Library Complementarity**:
   - Plotly provides interactive visualizations
   - Seaborn offers statistical visualization with high aesthetic quality
   - Both libraries have unique strengths for different visualization needs

3. **Data Insights**:
   - Life expectancy distributions vary significantly across continents
   - GDP and life expectancy show a positive correlation
   - Data density patterns are revealed through heatmaps and hexbin plots

4. **Visualization Enhancement**:
   - Marginal plots add context to main visualizations
   - Point overlays on violin plots show actual data distribution
   - Multiple visualization types can be combined for richer insights
